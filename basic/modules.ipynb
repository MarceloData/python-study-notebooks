#### First module

```python
open("mymath.py", "w").write(
    '''"""my math - our example math module"""
    pi = 3.14159
    def area(r):
    """area(r): return the area of a circle with redius r."""
    return(pi * r * r)
''')
```

#### Imports

```python
import mymath
print(mymath.pi)
print(mymath.area(2))
print(mymath.__doc__)
print(mymath.area.__doc__)
```

```python
from mymath import pi
pi
area(2)
```

```python
from mymath import *
print(pi)
print(area(2))
```
